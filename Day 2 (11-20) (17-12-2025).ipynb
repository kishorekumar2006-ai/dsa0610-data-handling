11. Area Plot
Dataset: Yearly Revenue

```r
# Load required library
library(ggplot2)

# Create the data frame
revenue_data <- data.frame(
  Year = 2015:2019,
  Revenue = c(30, 35, 40, 45, 50)
)

# Plot area chart
ggplot(revenue_data, aes(x = Year, y = Revenue)) +
  geom_area(
    fill = "skyblue",
    alpha = 0.5
  ) +
  labs(
    title = "Yearly Revenue",
    x = "Year",
    y = "Revenue (in million)"
  ) +
  theme_minimal()
```

12. Step Plot
Dataset: Cumulative Sales

```r
# Load required library
library(ggplot2)

# Create the data frame
cumulative_sales_data <- data.frame(
  Month = factor(
    c("Jan", "Feb", "Mar", "Apr", "May"),
    levels = c("Jan", "Feb", "Mar", "Apr", "May")
  ),
  Cumulative_Sales = c(100, 200, 300, 400, 500)
)

# Create the step plot
ggplot(cumulative_sales_data, aes(x = Month, y = Cumulative_Sales, group = 1)) +
  geom_step(linewidth = 1, color = "blue") +
  labs(
    title = "Cumulative Sales Over Months",
    x = "Month",
    y = "Cumulative Sales"
  ) +
  theme_minimal()
```

13. Ridgeline Plot
Dataset: Temperature Distributions

```r
install.packages("ggridges")
```

```r
# Load required libraries
library(ggplot2)
library(ggridges)

# Create the data frame
temperature_data <- data.frame(
  City = c("City1", "City1", "City1", "City2", "City2", "City2"),
  Temperature = c(20, 21, 19, 22, 23, 24)
)

# Plot ridgeline plot
ggplot(temperature_data, aes(x = Temperature, y = City, fill = City)) +
  geom_density_ridges(alpha = 0.7, scale = 1) +
  labs(
    title = "Temperature Distributions",
    x = "Temperature",
    y = "City"
  ) +
  theme_minimal()
```

14. Dumbbell Plot
Dataset: Sales Before and After Campaign

```r
# Load required library
library(ggplot2)

# Create the data frame
sales_campaign_data <- data.frame(
  Product = c("A", "B", "C"),
  Before = c(200, 300, 400),
  After  = c(250, 350, 450)
)

# Dumbbell plot using ggplot2 only
ggplot(sales_campaign_data) +
  geom_segment(
    aes(y = Product, yend = Product, x = Before, xend = After),
    color = "grey",
    linewidth = 1.2
  ) +
  geom_point(
    aes(x = Before, y = Product),
    color = "blue",
    size = 4
  ) +
  geom_point(
    aes(x = After, y = Product),
    color = "red",
    size = 4
  ) +
  labs(
    title = "Sales Before and After Campaign",
    x = "Sales",
    y = "Product"
  ) +
  theme_minimal()
```

15. Lollipop Plot
Dataset: Sales by Region

```r
# Load required library
library(ggplot2)

# Create the data frame
sales_region_data <- data.frame(
  Region = c("North", "South", "East", "West"),
  Sales = c(150, 200, 180, 210)
)

# Plot lollipop chart
ggplot(sales_region_data, aes(x = Region, y = Sales)) +
  geom_segment(
    aes(x = Region, xend = Region, y = 0, yend = Sales),
    color = "grey",
    linewidth = 1.2
  ) +
  geom_point(
    color = "blue",
    size = 5
  ) +
  labs(
    title = "Sales by Region",
    x = "Region",
    y = "Sales"
  ) +
  theme_minimal()
```

16. Spaghetti Plot
Dataset: Sales over Time by Product

```r
# Load required library
library(ggplot2)

# Create the data frame
sales_time_data <- data.frame(
  Month = factor(
    c("Jan", "Feb", "Mar", "Jan", "Feb", "Mar"),
    levels = c("Jan", "Feb", "Mar")
  ),
  Product = c("A", "A", "A", "B", "B", "B"),
  Sales = c(100, 150, 130, 120, 140, 160)
)

# Plot spaghetti plot
ggplot(sales_time_data, aes(x = Month, y = Sales, group = Product, color = Product)) +
  geom_line(linewidth = 1.2) +
  geom_point(size = 3) +
  labs(
    title = "Sales over Time by Product",
    x = "Month",
    y = "Sales"
  ) +
  theme_minimal()
```

17. Waterfall Plot
Dataset: Profit and Loss by Month

```r
# Install the package if not already installed
install.packages("waterfalls")

# Load required library
library(waterfalls)

# Create the data frame
profit_loss_data <- data.frame(
  labels = c("Jan", "Feb", "Mar", "Apr", "May"),
  values = c(100, -20, 50, -10, 80)
)

# Plot waterfall chart
waterfall(
  profit_loss_data,
  fill_by_sign = TRUE
)
```

18. Hexbin Plot

```r
# Install package if not already installed
install.packages("hexbin")

# Load required libraries
library(ggplot2)
library(hexbin)

# Create the data frame
xy_data <- data.frame(
  X = c(1, 2, 3, 4, 5, 6, 7, 8),
  Y = c(2, 3, 4, 5, 6, 7, 8, 9)
)

# Plot hexbin chart
ggplot(xy_data, aes(x = X, y = Y)) +
  geom_hex(bins = 5) +
  labs(
    title = "Hexbin Plot",
    x = "X",
    y = "Y"
  ) +
  theme_minimal()
```

20. Calendar Heatmap
Dataset: Daily Activity Count

```r
# Load required libraries
library(ggplot2)
library(lubridate)

# Create the data frame
activity_data <- data.frame(
  Date = as.Date(c(
    "2023-01-01",
    "2023-01-02",
    "2023-01-03",
    "2023-01-04",
    "2023-01-05"
  )),
  Count = c(10, 12, 15, 8, 20)
)

# Add weekday and week-of-year columns
activity_data$Weekday <- wday(activity_data$Date, label = TRUE)
activity_data$Week <- week(activity_data$Date)

# Plot calendar heatmap
ggplot(activity_data, aes(x = Weekday, y = Week, fill = Count)) +
  geom_tile(color = "white") +
  scale_fill_gradient(low = "white", high = "blue") +
  labs(
    title = "Daily Activity Count",
    x = "Weekday",
    y = "Week of Year",
    fill = "Count"
  ) +
  theme_minimal()
```
